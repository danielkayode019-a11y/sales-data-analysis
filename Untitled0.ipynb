```python
import pandas as pd

data = {
    "Employee": ["John", "Mary", "David", "Sarah", "Daniel"],
    "Department": ["IT", "HR", "IT", "Finance", "HR"],
    "Salary": [5000, 4500, 6000, 4000, 5500],
    "Years_Experience": [3, 5, 4, 2, 6]
}

df = pd.DataFrame(data)
# add bonus to salary
df["Bonus"] = df["Salary"] *0.10

# shows the total salary
df["Total_salary"] = df["Bonus"] + df["Salary"]

# shows worker with the higest salry
Highest_Paid = df.loc[df["Total_salary"].idxmax()]

# average salary in each department
department_salary = df.groupby ("Department")["Salary"].mean()

# experienced employees
experienced_employees = df[df["Years_Experience"]>4]

# sorting by salary
sorted_salary = df.sort_values(by = "Total_salary", ascending = False)

# Employee with the lowest salary
Lowest_salary = df.loc[df["Salary"].idxmin()]

#Total salary paid
Total_salary_with_bonus  = df["Total_salary"].sum()

# performance of employee
def check_performance(years):
  if years >= 5:
    return "Senior"
  else:
    return "Junior"

df["Performance"] = df ["Years_Experience"].apply(check_performance)

print(sorted_salary)
print("Highest_Paid")
print("Total_salary")
print(department_salary)
print(experienced_employees)
print(Lowest_salary)
print(Total_salary_with_bonus)
print(df)

#visualization
import matplotlib.pyplot as plt
plt.bar (df["Employee"], df["Salary"])
plt.title("Employee total salary")
plt.xlabel("Employee")
plt.ylabel("Salary")
plt.show()
```
